2. Выбор метрики оценивания модели.

Базовыми метриками оценивания являются precision, recall, f1-score, accuracy.

TP - истинно положительный: объект принадлежал определенному классу и модель правильно его предсказала.

FP - ложно положительный: объект не принадлежал определенному классу и модель предсказала его принадлежность к нему.

TN - истинно отрицательный: объект не принадлежал определенному классу и модель не предсказала его принадлежность к нему.

FN - ложно отрицательный: объект принадлежал определенному классу и модель не предсказала его принадлежность к нему.

Метрика accuracy: так как классы имеют очень серьезный дисбаланс, применять accuracy некорректно, так как результат вероятнее всего будет недостоверным.

Метрика precision: исходя из формул, приведенных ниже, это доля верных правильных предсказаний среди всех предсказанных к этому классу объектов. 

Метрика recall: доля верных правильных предсказаний среди всех предсказаний, реально относящихся к этому классу объекта.

Метрика f1-score: гармоническое среднее, показывающее общую картину, универсальная метрика.

Применительно к моей задаче: 

FN - атака была, но модель ее не распознала, это приведет к значительному ущербу.

FP - модель ошиблась, но атаку заметили, это приводит к ложному срабатыванию, но ущерба нет.

Таким образом, необходимо ориентироваться на метрику recall.

3.1. KNN - алгоритм машинного обучения, который для каждого нового объекта смотрит на k ближайщих соседей, при классификации присваивает ему класс большинства среди этих соседей.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
from sklearn.utils import shuffle
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3, p=1)
model_knn.fit(X_train, y_train)
y_pred_knn = model_knn.predict(X_test)
```

```python
conf_mat_knn = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_mat_knn, annot=True, fmt='d', cmap='Blues')
plt.title('Матрица ошибок для KNN')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

print(classification_report(y_test, y_pred_knn))
```

Так как классы 4, 5, 8 встречаются крайне редко, то можно ими пренебречь при рассчете метрики.

```python
from sklearn.metrics import recall_score
classes = [1.0, 2.0, 3.0, 6.0, 7.0, 9.0]
mean_recall_knn = recall_score(
    y_test, y_pred_knn, 
    labels = classes, 
    average = 'macro'
)
print(f"Средний recall по необходимым классам: {mean_recall_knn:.3f}")
```

Получила уже достаточно неплохой резульат, но попробую поэкспериментировать с параметрами.

```python
from sklearn.model_selection import GridSearchCV

model_knn = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 15],
    'p': [1, 2]
}

grid_search = GridSearchCV(
    estimator = model_knn,
    param_grid = param_grid,
    cv = 5,
    scoring = 'recall_macro',
    n_jobs = -1
)

grid_search.fit(X_train, y_train)

best_knn = grid_search.best_estimator_

print("Лучшие параметры:", grid_search.best_params_)
print("Лучший recall:", grid_search.best_score_)
```

Оказалось, что параметры изначально были подобраны корректно. Нормализую данные для того, чтобы привести признаки к единому масштабу.
